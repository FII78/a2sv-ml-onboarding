# **Parsing the .csv file into a data frame**


```python
import pandas as pd

# Read the CSV file into a data frame
data_frame = pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')

# Now you can work with the data frame
# For example, you can print the first few rows
print(data_frame.head())
```

# **What are the column names in the dataset?**

```python
column_names = data_frame.columns
```

```python
column_names = data_frame.columns.tolist()
print(column_names)
```

# **What is the average rating of the movies?**

```python
average_rating = data_frame['IMDB_Rating'].mean()

# Print the average rating
print("Average Rating:", average_rating)
```

# **How many unique genres are there?**

```python
# Extract the unique genres
unique_genres = data_frame['Genre'].unique()
# Count the number of unique genres
num_unique_genres = len(unique_genres)

# Print the number of unique genres
print("Number of Unique Genres:", num_unique_genres)
```

```python
# Create a DataFrame with the answers
answers_df = pd.DataFrame({'Question': ['Column names', 'Average rating', 'Num unique genres'],
                           'Answer': [column_names, average_rating, num_unique_genres]})

# Save the DataFrame as a CSV file
answers_df.to_csv('answers_formatted.csv', index=False)

# Print the DataFrame
print(answers_df)
```

# **Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list.**

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.hist(data_frame['Genre'], bins=10)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Distribution of Movies')
plt.xticks(rotation=90)
plt.show()
```

# **Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list)**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the number of movies for each year
movies_per_year = data_frame['Released_Year'].value_counts().sort_index()

# Create the scatterplot
plt.scatter(movies_per_year.index, movies_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies vs Year')
plt.show()
```
